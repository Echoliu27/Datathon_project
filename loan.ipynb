```python
import pandas as pd
import numpy as np
```

```python
user_cluster = pd.read_csv('~/Downloads/user_categorical_w_cluster.csv')
user_apply_loan = pd.read_csv('~/Downloads/df_loans.csv')
```

```python
user1 = pd.merge(left = user_cluster,right=user_apply_loan,left_on='user_id',right_on='user_id',how='inner')
```

```python
user1 = user1.drop(['user_id','user_signup_timestamp','zipcode','gender','recent_bankruptcy_date','Unnamed..0'],1)
user1
```

```python
#convert categorical into dummy
user_dummy1 = pd.get_dummies(user1,drop_first=True,columns=['state','is_homeowner',
                                                          'age_bucket','credit_score_bucket','cluster','applied'])
user_dummy1
```

```python
#sample
df_sample1 = user_dummy1.sample(frac = 0.1, replace = False)
df_sample1
```

```python
#split
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test=train_test_split(df_sample1.drop('applied_Y',1),list(df_sample1['applied_Y']),test_size=0.3,random_state=8)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import RidgeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
param_test1 = {'learning_rate':[0.01,0.05,0.1],'n_estimators':range(20,81,10)}
gsearch2 = GridSearchCV(estimator = GradientBoostingClassifier(min_samples_split=300,
                                  min_samples_leaf=20,max_depth=8,max_features='sqrt', subsample=0.8,random_state=10), 
                       param_grid = param_test1, scoring='roc_auc',iid=False,cv=5)
gsearch2.fit(x1_train,y1_train)
print('Test accuracy: %.3f' % gsearch2.score(x1_test, y1_test))
```

```python
gsearch2.best_estimator_
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
y_predict = gsearch2.predict(x1_test)
conf_mat = confusion_matrix(y1_test,y1_predict)
conf_mat
```

```python
tp1,fp1,fn1,tn1=conf_mat.ravel()
accuracy = (tp1+tn1)/(tp1+fp1+fn1+tn1)
sensitivity = tp1/(tp1+fn1) #recall
specifity = tn1/(fp1+tn1)
PPV = tp1/(tp1+fp1) #precision
NPV = tn1/(fn1+tn1)
F1 = (2*tp1)/(2*tp1+fn1+fp1)
print('accuracy =',accuracy)
print('sensitivity =',sensitivity)
print('specifity =',specifity)
print('PPV =',PPV)
print('NPV =',NPV)
print('F1 score =',F1)
```
